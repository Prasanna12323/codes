Problem Statement: Implement perceptron learning law with its decision regions using python. Give the output in graphical form 

```python
# This cell contains ANN-related code
# Comments are added for understanding
import numpy as np  # Import libraries
import matplotlib.pyplot as plt  # Import libraries
```

```python
# This cell contains ANN-related code
# Comments are added for understanding
# 🔹 Perceptron Class
class Perceptron:
    def __init__(self, lr=0.1, epochs=100):  # Assign value
        self.lr = lr  # Assign value
        self.epochs = epochs  # Assign value
        
        # initialize weights
        self.w1 = 0  # Assign value
        self.w2 = 0  # Assign value
        self.b = 0  # Assign value

    # 🔹 Activation Function
    def activation(self, z):
        if z >= 0:  # Assign value
            return 1
        else:
            return -1

    # 🔹 Training Function
    def fit(self, X, y):  # Train model
        for epoch in range(self.epochs):
            for i in range(len(X)):
                
                x1 = X[i][0]  # Assign value
                x2 = X[i][1]  # Assign value
                target = y[i]  # Assign value
                
                # weighted sum
                z = (self.w1 * x1) + (self.w2 * x2) + self.b  # Assign value
                
                # prediction  # Predict output
                y_pred = self.activation(z)  # Assign value
                
                # update if wrong
                if y_pred != target:  # Assign value
                    self.w1 += self.lr * target * x1  # Assign value
                    self.w2 += self.lr * target * x2  # Assign value
                    self.b += self.lr * target  # Assign value

    # 🔹 Prediction Function
    def predict(self, x1, x2):  # Predict output
        z = (self.w1 * x1) + (self.w2 * x2) + self.b  # Assign value
        return self.activation(z)
```

```python
# This cell contains ANN-related code
# Comments are added for understanding
# 🔹 Function to Plot Decision Regions
def plot_decision_boundary(model, X, y):
    
    x_values = np.linspace(0, 8, 100)  # Assign value
    y_values = np.linspace(0, 8, 100)  # Assign value
    
    xx, yy = np.meshgrid(x_values, y_values)  # Assign value
    
    Z = model.w1 * xx + model.w2 * yy + model.b  # Assign value
    Z = np.where(Z >= 0, 1, -1)  # Assign value
    
    plt.contourf(xx, yy, Z, alpha=0.3)  # Assign value
    
    # plot points
    for i in range(len(X)):
        if y[i] == 1:  # Assign value
            plt.scatter(X[i][0], X[i][1], marker='o', label='Class 1' if i == 3 else "")  # Assign value
        else:
            plt.scatter(X[i][0], X[i][1], marker='x', label='Class -1' if i == 0 else "")  # Assign value
    
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.title("Perceptron Decision Regions")
    plt.legend()
    plt.show()
```

```python
# This cell contains ANN-related code
# Comments are added for understanding
# 🔹 Dataset
X = np.array([  # Assign value
    [2, 3],
    [4, 5],
    [1, 1],
    [2, 1],
    [6, 7],
    [7, 8]
])
```

```python
# This cell contains ANN-related code
# Comments are added for understanding
y = np.array([-1, -1, -1, 1, 1, 1])  # Assign value
```

```python
# This cell contains ANN-related code
# Comments are added for understanding
# 🔹 Create Model
model = Perceptron(lr=0.1, epochs=100)  # Assign value
```

```python
# This cell contains ANN-related code
# Comments are added for understanding
# 🔹 Train Model
model.fit(X, y)  # Train model
```

```python
# This cell contains ANN-related code
# Comments are added for understanding
# 🔹 Print Results
print("Final Weights:", model.w1, model.w2)
print("Final Bias:", model.b)
```

```python
# This cell contains ANN-related code
# Comments are added for understanding
# 🔹 Plot
plot_decision_boundary(model, X, y)
```
