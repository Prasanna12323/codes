```python
# This cell contains code for an Artificial Neural Network (ANN)
# Comments are added to help you understand each step
import numpy as np  # Import required libraries
import matplotlib.pyplot as plt  # Import required libraries
```

```python
# This cell contains code for an Artificial Neural Network (ANN)
# Comments are added to help you understand each step
x = np.linspace(-10, 10, 100)  # Variable assignment
```

```python
# This cell contains code for an Artificial Neural Network (ANN)
# Comments are added to help you understand each step
def plot_activation(name, x, y):
    plt.figure(figsize=(8, 4))  # Variable assignment
    plt.plot(x, y, color='blue', linewidth=2)  # Variable assignment
    plt.title(f"{name} Activation Function")
    plt.axhline(0, color='black', lw=1, alpha=0.5)  # Variable assignment
    plt.axvline(0, color='black', lw=1, alpha=0.5)  # Variable assignment
    plt.grid(True, linestyle='--', alpha=0.7)  # Variable assignment
    plt.show()
```

```python
# This cell contains code for an Artificial Neural Network (ANN)
# Comments are added to help you understand each step
def step(x):
    return np.where(x >= 0, 1, 0)  # Variable assignment
```

```python
# This cell contains code for an Artificial Neural Network (ANN)
# Comments are added to help you understand each step
print("Plotting Step Function...")
plot_activation("Step", x, step(x))
```

```python
# This cell contains code for an Artificial Neural Network (ANN)
# Comments are added to help you understand each step
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
# This cell contains code for an Artificial Neural Network (ANN)
# Comments are added to help you understand each step
print("Plotting Sigmoid...")
plot_activation("Sigmoid", x, sigmoid(x))
```

```python
# This cell contains code for an Artificial Neural Network (ANN)
# Comments are added to help you understand each step
def bipolar_sigmoid(x):
    return np.tanh(x)
```

```python
# This cell contains code for an Artificial Neural Network (ANN)
# Comments are added to help you understand each step
print("Plotting Bipolar Sigmoid (Tanh)...")
plot_activation("Bipolar Sigmoid", x, bipolar_sigmoid(x))
```

```python
# This cell contains code for an Artificial Neural Network (ANN)
# Comments are added to help you understand each step
def relu(x):
    return np.maximum(0, x)
```

```python
# This cell contains code for an Artificial Neural Network (ANN)
# Comments are added to help you understand each step
print("Plotting ReLU...")
plot_activation("ReLU", x, relu(x))
```

```python
# This cell contains code for an Artificial Neural Network (ANN)
# Comments are added to help you understand each step
def leaky_relu(x):
    return np.where(x > 0, x, 0.01 * x)
```

```python
# This cell contains code for an Artificial Neural Network (ANN)
# Comments are added to help you understand each step
print("Plotting Leaky ReLU...")
plot_activation("Leaky ReLU", x, leaky_relu(x))
```

```python
# This cell contains code for an Artificial Neural Network (ANN)
# Comments are added to help you understand each step
def elu(x, alpha=1):  # Variable assignment
    return np.where(x > 0, x, alpha * (np.exp(x) - 1))
```

```python
# This cell contains code for an Artificial Neural Network (ANN)
# Comments are added to help you understand each step
print("Plotting ELU...")
plot_activation("ELU", x, elu(x))
```

```python
# This cell contains code for an Artificial Neural Network (ANN)
# Comments are added to help you understand each step
def softmax(x):
    e_x = np.exp(x - np.max(x))  # Variable assignment
    return e_x / e_x.sum()
```

```python
# This cell contains code for an Artificial Neural Network (ANN)
# Comments are added to help you understand each step
sample_inputs = np.array([2.0, 1.0, 0.1, -1.0, 3.5])  # Variable assignment
soft_outputs = softmax(sample_inputs)  # Variable assignment
```

```python
# This cell contains code for an Artificial Neural Network (ANN)
# Comments are added to help you understand each step
plt.figure(figsize=(8, 4))  # Variable assignment
plt.bar(range(len(sample_inputs)), soft_outputs, color='purple', alpha=0.7)  # Variable assignment
plt.xticks(range(len(sample_inputs)), sample_inputs)
plt.title("Softmax Output (Probabilities for specific inputs)")
plt.xlabel("Input Value")
plt.ylabel("Probability")
plt.show()
```
